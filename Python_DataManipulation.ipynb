```python
import pandas as pd
```

```python
#Merging Data Frames
max_weather = pd.read_csv('sample_data/monthly_max_temp.csv', index_col = 'Month')
mean_weather = pd.read_csv('sample_data/monthly_mean_temp.csv', index_col = 'Month')

weather_1 = max_weather.copy()
weather_1
```

```python
weather_2 = mean_weather.copy()
weather_2
```

```python
weather_1.index
```

```python
#Sorting Index
weather2= weather_1.sort_index()
weather2

```

```python
#Sorting Index in reverse order
weather3=weather_1.sort_index(ascending = False)
weather3
```

```python
#Sorting Index using the Max Temperature
weather4 = weather_1.sort_values('Max TemperatureF')
weather4
```

```python
#Reindexing
ordered = ['Oct','Jan','May','Apr']
weather_1.reindex(ordered)
```

```python
#Reindexing using another dataframe
weather_2.index
```

```python
weather_1.reindex(weather_2.index)
```

```python
#Reindexing using a value which does not exsist in the first dataframe
order = ['Jan', 'Apr', 'XYZ']
weather_1.reindex(order)
```

```python
#dropping the NA value
order = ['Jan', 'Apr', 'XYZ']
weather_1.reindex(order).dropna()
```

```python
#Fillling the NA value
order = ['Jan', 'Apr', 'XYZ']
weather_1.reindex(order).ffill()
weather_1.reindex(order).bfill()
```
